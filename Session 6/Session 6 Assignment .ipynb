```python
# Que 1
def two_sum_closest(nums, target):
    if not nums or len(nums) < 2:
        return None

    left, right = 0, len(nums) - 1
    closest_sum = float('inf')
    result = None

    while left < right:
        current_sum = nums[left] + nums[right]

        if abs(current_sum - target) < abs(closest_sum - target):
            closest_sum = current_sum
            result = (nums[left], nums[right])

        if current_sum < target:
            left += 1
        else:
            right -= 1

    return result

# Example usage:
sorted_nums = [1, 2, 3, 4, 5, 6, 7]
target = 8
closest_pair = two_sum_closest(sorted_nums, target)
print(closest_pair) 
```

```python
#Que 2

def max_profit(prices):
    if not prices:
        return 0

    n = len(prices)
    dp = [0] * n

    for i in range(1, n):
        max_profit = dp[i - 1]  # Maximum profit if we don't buy/sell on the i-th day

        # Try selling the stock on the i-th day and buying on any previous day
        for j in range(i):
            max_profit = max(max_profit, prices[i] - prices[j] + (dp[j - 2] if j >= 2 else 0))

        dp[i] = max_profit

    return dp[-1]

# Example usage:
prices = [1, 2, 3, 0, 2]
profit = max_profit(prices)
print(profit)  # Output: 3 (Buy on day 1 (price = 1) and sell on day 3 (price = 3), buy on day 4 (price = 0) and sell on day 5 (price = 2))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def find_twin_primes(n):
    twin_primes = []
    candidate = 3  # Start with the first possible twin prime

    while len(twin_primes) < n:
        if is_prime(candidate) and is_prime(candidate - 2):
            twin_primes.append((candidate - 2, candidate))
        candidate += 2  # Only consider odd numbers

    return twin_primes

# Find the first 100 twin primes
first_100_twin_primes = find_twin_primes(100)

# Print the first 100 twin prime pairs
for twin_prime in first_100_twin_primes:
    print(twin_prime)
```
